# Distribución uniforme y binomial

## Simulación distribución uniforme

```r
runif(5)
options(repr.plot.width=6, repr.plot.height=6/1.61803)
hist(runif(5),main="Distribución uniforme 5 números",col="blue")

set.seed(600)
options(repr.plot.width=6, repr.plot.height=6/1.61803)
hist(runif(50), main="Distribución uniforme 50 números", col="red")
```

```r
rbinom(n=30,size=1,prob=0.5)
volados30 <- rbinom(n=30,size=1,prob=0.5) 
layout(matrix(1:3,1,3))
hist(volados30)
volados300 <- rbinom(n=300,size=1,prob=0.5)
hist(volados300)
volados3millones <- rbinom(n=3000000,size=1,prob=0.5)
hist(volados3millones)
```

De la gráfica se aprecia como cuando n tiende a infinito las probabilidades se asemejan más a 0.5
