# TEOREMA DEL LIMITE CENTRAL

A fin de tener todos la misma semilla para el ejecicio, se fija la semilla para generar los números aleatorios

```r
set.seed(555)
```

Se hace una simulación con 50 datos

```r
sim50 <- rnorm(50,0,1)
```

```r
sim50
```

Se contrasta con una simulacion para 5 millones de datos

```r
sim5MM <- rnorm(5000000,0,1)
```

```r
options(repr.plot.width=6,repr.plot.height=6/1.61803)
hist(sim50,main="SIMULACION DE 50 NUMEROS ALEATORIOS",xlab="VALORES ALEATORIOS",ylab="FRECUENCIA ABSOLUTA",ylim= c(0,15),col="green")
```

```r
options(repr.plot.width=6,repr.plot.height=6/1.61803)
hist(sim5MM, main="SIMULACION DE 5 MILLONES NUMEROS ALEATORIOS",xlab="VALORES ALEATORIOS",ylab="FRECUENCIA ABSOLUTA",ylim = c(0,1000000),col="green")
```

La comparación se muestra mejor en la siguiente matriz

```r
layout(matrix(1:2,1,2))
hist(sim50,main="SIM 50 NUM. ALEATORIOS",xlab="VALORES ALEATORIOS",ylab="FRECUENCIA ABSOLUTA",ylim= c(0,25),col="green")
hist(sim5MM, main="SIM DE 5MM  NUM ALEATORIOS",xlab="VALORES ALEATORIOS",ylab="FRECUENCIA ABSOLUTA",ylim = c(0,1000000),col="green")
```

Los estadísticos de las simulaciones se muestran a continuación

```r
summary (sim50)
```

```r
summary (sim5MM)
```

```r
mean (sim50)
```

```r
mean (sim5MM)
```

```r
dif<- 1-(abs(mean(sim50))-abs(mean(sim5M)))/abs(mean(sim50))
```

```r
dif*100
```

Esto significa que la diferencia es despreciable y por tanto con muestras a partir de 30 se considera que se tienen valores suficientes para calcular los parámetros de la población.
